## Objective

To explore and play with the Minimum Wage Dataset.

```python
import pandas as pd
```

```python
# Import the dataset
df = pd.read_csv('datasets/Minimum Wage Data.csv')

# If utf-8 encoded data isn't available by default, use latin encoding
df = pd.read_csv('datasets/Minimum Wage Data.csv', encoding='latin')
# And, save the UTF8 dataset to use directly in future
df.to_csv('datasets/minwage.csv', encoding='utf-8')
```

```python
df
```

```python
# Create groupby objects 
gb = df.groupby('State')

# To grab a group
gb.get_group('Alabama').set_index('Year').head()
```

```python
# Iterate over a group
act_min_wage = pd.DataFrame()

# Here name is the value on which dataset is grouped (Alaska, etc)
# And, group is the actual groupby object for that name
for name, group in df.groupby('State'):
    if act_min_wage.empty:
        act_min_wage = group.set_index('Year')[['Department.Of.Labor.Cleaned.Low.Value.2020.Dollars']].rename(columns={'Department.Of.Labor.Cleaned.Low.Value.2020.Dollars': name})
    else:
        act_min_wage = act_min_wage.join(group.set_index('Year')[['Department.Of.Labor.Cleaned.Low.Value.2020.Dollars']].rename(columns={'Department.Of.Labor.Cleaned.Low.Value.2020.Dollars': name}))
    
act_min_wage.head()
```

```python
# Get a quick rundown of basis statistics of the dataset 
act_min_wage.describe()
```

```python
# Get the Coorelation matrix
act_min_wage.corr()
```

```python
# Identify the reason behind NaN values in above correlation matrix
# While importing csv, the blank values are read as 0s
issue_df = df[df['Department.Of.Labor.Cleaned.Low.Value.2020.Dollars'] == 0]
issue_df
```

```python
# Identify States we can't get data on
issue_df['State'].unique()
```

```python
# Remove the States with missing data
import numpy as np

# Axis = 1 refers to columns ; Axis = 0 is for rows
act_min_wage.replace(0, np.NaN).dropna(axis=1).corr()
```

```python
# Upon removing missing values, we notice that respective State columns are dropped from correlation matrix
min_wage_corr = act_min_wage.replace(0, np.NaN).dropna(axis=1).corr().head()

for problem in issue_df['State'].unique():
    if problem in min_wage_corr.columns:
        print('We are missing something here...')
```

```python
# Often it is not the right approach to remove the entire column if there's a single NaN value in it
grouped_issues = issue_df.groupby('State')
grouped_issues.get_group('Alabama').head()
```

```python
# Iterate over the grouped data and check if entire column has missing data
grouped_issues.get_group('Alabama')['Department.Of.Labor.Cleaned.Low.Value.2020.Dollars'].sum()
```

```python
# If the entire column adds up to 0, we never got Min Wage data for that State
for state, data in grouped_issues:
    if data['Department.Of.Labor.Cleaned.Low.Value.2020.Dollars'].sum() != 0:
        print('We missed something!')
```
